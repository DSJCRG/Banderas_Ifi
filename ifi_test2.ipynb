```python
import pandas as pd 
import pyodbc
import sqlalchemy
import datetime as dt 
import numpy as np

#Se definen las conexiones

def get_engine():
    connection_string = "mssql+pyodbc://{0}@{2}:{1}@{2}.database.windows.net:{4}/{3}?driver=ODBC+Driver+17+for+SQL+Server".format(
                        'jrivas',
                        'XXXXX',
                        'koice',
                        'ko_ice',
                        1433)
    res = sqlalchemy.engine.create_engine(connection_string, connect_args={'autocommit': True})
    return res
    
def get_query(query):
    engine = get_engine()
    res = pd.read_sql(query, con=engine)
    return res
```

```python
query = """ 


select distinct tt.user_id, tt.date_id, tt.is_reported, sl.Question, sl.Response, rpt.YearNbr, rpt.MonthDsc, rpt.DateFormat, rpt.LocationId, rpt.Code, rpt.BottlerName, rpt.LocationClusterName, rpt.MarketSegment, rpt.Segment, rpt.SessionUId, rpt.SceneUId, rpt.SceneType, rpt.SubsceneType, rpt.UserName, rpt.UserRole, rpt.Enf_KO, rpt.Enf_Comp, rpt.Enf_Propio 
from dbo.RPT_IR_TT_AllBottlers rpt 
join dbo.SurveyList sl on sl.SessionUid = rpt.SessionUId
join dbo.a_tt_10_CMP_r_traditional_data_01 tt on tt.SessionUid = sl.SessionUid
where rpt.YearNbr = 2021 and rpt.LocationClusterName = 'Traditional' and rpt.UserRole = 'Auditor' and sl.Question = '¿La tienda tiene libre acceso a Clientes?' and rpt.SceneType = 'Enfriador' 

 
"""
```

```python
df = pd.DataFrame(get_query(query))
#df = df.sort_values(by="VisitStartDateTime",ascending=False)
df
```

```python
df.duplicated(subset='SessionUId', keep='first').sum()
```

```python
df=df.drop_duplicates(subset=['SessionUId'])
df
```

```python
df[["DateFormat"]] = df[["DateFormat"]].apply(pd.to_datetime)
```

```python
df.info()
```

```python
df['day'] = df['DateFormat'].dt.day

```

```python
df.info()
```

```python
df['month'] = df['DateFormat'].dt.month
df.info()
```

```python
table = pd.pivot_table(df, values=['Enf_KO','Enf_Comp', 'Enf_Propio'], index=['user_id', 'date_id', 'LocationId', 'is_reported', 'DateFormat', 'month', 'day', 'SessionUId'],
                     fill_value=0)
table=pd.DataFrame(table)
table
```

```python
table = table.reset_index()
table
```

```python
table.info()
```

```python
moni=table.groupby(['user_id', 'date_id', 'LocationId', 'SessionUId', 'is_reported', 'DateFormat', 'month', 'day'])['Enf_KO', 'Enf_Comp', 'Enf_Propio'].sum()
moni
```

```python
table2 = moni.reset_index()
table2
```

```python
table2=table2.sort_values(['LocationId', 'DateFormat'])
table2
```

```python
table2["Coca-Cola"] = table2["Enf_KO"].shift(1)
table2["Competencia"] = table2["Enf_Comp"].shift(1)
table2["Propio"] = table2["Enf_Propio"].shift(1)
table2.dtypes
```

```python
table2 = table2.fillna(0)
```

```python
table2 = table2.astype({"Coca-Cola": int, "Competencia": int, "Propio": int, "LocationId": object})
```

```python
table2.info()
```

```python
table2["LocationIdChanged"] = table2["LocationId"].shift(1, fill_value=table2["LocationId"].head(1)) != table2["LocationId"]
table2.dtypes
```

```python
table2["LocationIdChanged"] = table2["LocationIdChanged"].astype(object)
table2.dtypes
```

```python
for OutletCodeChanged in table2:
      s = table2['LocationIdChanged'].unique()
      for i in s:
          if i == 'True':
              Flag_Prop = 'change'
              
          else: 
              Flag_Prop = table2["Enf_Propio"] - table2["Propio"]

              table2['Flag_Prop']=Flag_Prop
```
